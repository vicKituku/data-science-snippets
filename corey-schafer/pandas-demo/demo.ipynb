```python
country_respondents = df_SO_data['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('JavaScript').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis = 'columns', sort=False)
```

```python
python_df.rename(columns={'count': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df.sort_values(by='PctKnowsPython',ascending=False).head(50)
python_df.loc['China']['PctKnowsPython']
```
