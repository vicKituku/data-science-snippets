## UNIVARIATE, BIVARIATE AND MULTIVARIATE ANALYSIS USING VISUALIZATION

```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load data
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
df['species'].value_counts()
```

## Univariate analysis

```python
df_setosa = df.loc[df['species'] == 'setosa']
df_versicolor = df.loc[df['species'] == 'versicolor']
df_virginica = df.loc[df['species'] == 'virginica']
```

```python
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']), marker = 'o', label = 'setosa')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']), marker = 'o', label = 'virginica')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), marker = 'o', label = 'versicolor')
plt.xlabel('Petal Length')
plt.legend()
plt.show()
```

# Bivariate analysis

```python
plt.figure(figsize=(7,7))
sns.scatterplot(data=df, x = 'sepal_length', y = 'sepal_width', hue='species', size='petal_length')
```

# Multivariate analysis

```python
sns.pairplot(df, hue='species',  diag_kind = "hist")
```
