```python
def fill_missing_with_median(df):
    # Selecting only the numerical columns
    numerical_columns = df.select_dtypes(include=['number']).columns
    print(numerical_columns)
    
    # Fill missing values with median for numerical columns
    df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
    
    return df
fill_missing_with_median(df)
```

```python
def drop_columns_with_missing_values(df, threshold):
    # Calculate the number of missing values in each column
    missing_values_count = df.isnull().sum()
    
    # Identify columns with missing values exceeding the threshold
    columns_to_drop = missing_values_count[missing_values_count > threshold].index
    
    # Drop the identified columns
    df = df.drop(columns=columns_to_drop)
    
    return df
drop_columns_with_missing_values(df,1200)
df

```
