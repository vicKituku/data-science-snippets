SEABORN FOR VISUALIZATION

```python
import seaborn as sns
```

```python
# datasets provided by seaborn
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
tips
```

SCATTER PLOT OF TIP AND TOTAL BILL

```python
sns.scatterplot(data = tips, x='tip', y = 'total_bill', hue='day', size = 'size', palette = 'YlGnBu')
```

HISTOGRAM PLOT

```python
sns.histplot(tips['tip'], kde = True, bins=30)
```

DISTRIBUTION PLOT

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(tips, x='sex', y='tip', palette='YlGnBu')
```

BOXPLOT

```python
sns.boxplot(tips, x= 'day', y = 'total_bill', hue = 'sex', palette = 'YlGnBu')
```

STRIPPLOT

```python
sns.stripplot(tips, x = 'day', y = 'total_bill', hue = 'sex', palette = 'YlGnBu', dodge = True)
```

JOINTPLOT

```python
sns.jointplot(tips, x='tip', y ='total_bill', kind = 'hex', cmap = 'YlGnBu')
```

PAIRPLOT

```python
sns.pairplot(titanic.select_dtypes(['number']), hue="pclass")
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap="YlGnBu")
```
